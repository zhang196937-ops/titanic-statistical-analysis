## Seaborn Plotting functions are divided into three systems./绘图函数分为三个层级: 
### Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.

## Table 1: Seaborn Graphical Hierarchy/Seaborn 图形层级结构

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# set styple 
sns.set_theme(style="whitegrid")
#loading dataset tips, tips is a pandas DataFrame.
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

## why whisker
    ================================================================================
    为什么用 1.5 × IQR 而不是简单的 min/max？
    ================================================================================
    
    1. 1.5 × IQR 的统计学依据：
       - 在正态分布下，这个范围大约覆盖 99.3% 的数据
       - 只有约 0.7% 的正常数据会被误判为异常值
       - 这是一个经验性的平衡点，既能识别异常值，又不会过度排除
    
    2. 为什么需要 whisker：
       - 如果直接使用 min/max，一个异常值就能把整个"箱子"拉长
       - whisker 代表"非异常值"的范围，更能反映数据的典型情况
       - 异常值被单独标记出来，既保留了信息，又不影响主体显示
    
    3. 业务价值：
       - 薪资分析：排除高管和实习生，看清普通员工薪资
       - 房价分析：排除豪宅和危房，看清普通房价
       - 质量控制：识别异常产品，监控生产过程
       - 医疗数据：识别异常指标，辅助诊断
    
    4. 可视化优势：
       - 箱体代表 50% 的数据
       - whisker 代表另外约 49.3% 的非异常数据
       - 异常值单独显示，完整呈现数据全貌
       - 既展示了主体分布，又不丢失异常值信息
    
    5. 1.5 这个系数的由来：
       - 统计学家 John Tukey 在 1977 年提出
       - 基于大量实际数据的经验总结
       - 在各种分布下都表现良好
       - 已经成为行业标准

## The meaning of each component in a boxplot

|组件|公式|含义|占数据比例|
|-|-|-|-|
|Q1|25%|分位数|下四分位数|25% 的数据 ≤ 此值|
|Q3|75%|分位数|上四分位数|75% 的数据 ≤ 此值|
|IQR|Q3 - Q1|四分位距|中间50% 数据的范围|
|下栅栏 lower fence|	Q1 - 1.5×IQR|异常值判断下界|低于此值为异常|
|上栅栏 upper fence|	Q3 + 1.5×IQR|异常值判断上界|高于此值为异常|
|下胡须 lower whisker|min(数据 ≥ 下栅栏)|正常值下限|约 24.65% 的数据|
|上胡须 upper whisker|max(数据 ≤ 上栅栏)|正常值上限|约 24.65% 的数据|
|异常值 outliers|data/数据 < lower fence/下栅栏 or/或 >upper fence/上栅栏	|特殊情况|约 0.7% 的数据|

```python
#add label for five-number summary (min, Q1, median, Q3, max) 
# method 1

plt.figure(figsize=(12, 6))
ax = sns.boxplot(
    data=tips, 
    x='day', 
    y='total_bill', 
    width=0.5  # define the width of the box.
)

# get unique value for day.
categories = tips['day'].unique()

for i, category in enumerate(categories):
    # get value for every day.
    # tips['day'] == category is filter and return a dataframe with all the column for every day.
    # data is a pandas series.
    # data is a one-dimensional array holding data of total_bill.
    data = tips[tips['day'] == category]['total_bill']
    
    # calulate 
    q1 = data.quantile(0.25)
    median = data.median()
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    
    # calculate whisker (实际最小/最大值，非异常值）
    lower_whisker = data[data >= lower_fence].min()
    upper_whisker = data[data <= upper_fence].max()
    
    # add labels.
    stats = [
        ('min', lower_whisker),
        ('Q1', q1),
        ('median', median),
        ('Q3', q3),
        ('max', upper_whisker)
    ]
    
    for label, value in stats:
        # 在数据点旁边添加文本
        ax.text(i, value, f'{label}\n{value:.1f}', 
                horizontalalignment='center',
                verticalalignment='bottom' if label in ['min', 'Q1', 'median'] else 'top',
                fontsize=8,
                color='red',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

plt.title("Boxplot with Labels")
plt.tight_layout()
plt.show()
```

```python
#add label for five-number summary (min, Q1, median, Q3, max) 
# method 2

fig, ax = plt.subplots(figsize=(14, 7))

# create boxplot.
sns.boxplot(
    data=tips, 
    x='day', 
    y='total_bill', 
    ax=ax, 
    width=0.4
)


categories = tips['day'].unique()
x_positions = range(len(categories))

for i, category in enumerate(categories):
    data = tips[tips['day'] == category]['total_bill']
    
    # calulate values
    stats = {
        'Min': data.min(),
        'Q1': data.quantile(0.25),
        'Median': data.median(),
        'Q3': data.quantile(0.75),
        'Max': data.max(),
        'IQR': data.quantile(0.75) - data.quantile(0.25)
    }
    
    # create label text.
    label_text = '\n'.join([f'{k}: {v:.2f}' for k, v in stats.items()])
    
    # add label.
    ax.text(i + 0.25, data.median(), label_text,
            fontsize=9,
            verticalalignment='center',
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

# Adjust the x-axis range to accommodate the label.
ax.set_xlim(-0.5, len(categories) - 0.5)

plt.title("Boxplot with Detailed Statistics")
plt.tight_layout()
plt.show()
```

```python
#add label for five-number summary (min, Q1, median, Q3, max) 
# method 3

fig, ax = plt.subplots(figsize=(16, 8))

# create chat.
sns.boxplot(data=tips, x='day', y='total_bill', ax=ax, width=0.5)

# add label.
categories = tips['day'].unique()

for i, category in enumerate(categories):
    data = tips[tips['day'] == category]['total_bill']
    
    q1 = data.quantile(0.25)
    median = data.median()
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    
    lower_whisker = max(data.min(), q1 - 1.5 * iqr)
    upper_whisker = min(data.max(), q3 + 1.5 * iqr)
    
    points = [
        ('Min', lower_whisker, 'bottom'),
        ('Q1', q1, 'bottom'),
        ('Median', median, 'center'),
        ('Q3', q3, 'top'),
        ('Max', upper_whisker, 'top')
    ]
    
    for label, value, va in points:
        offset = 0.15 if label in ['Min', 'Q1'] else -0.15
        ax.plot([i, i + offset], [value, value], 'k--', alpha=0.5, linewidth=1)
        ax.text(i + offset, value, f' {label}: {value:.1f} ',
                fontsize=9, verticalalignment=va,
                horizontalalignment='left' if offset > 0 else 'right',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='gray', alpha=0.9))

plt.title("Boxplot with Annotated Statistics")#, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Violin Plot / 小提琴图
#Combines box plot with kernel density estimation to show full distribution shape.
#结合箱线图和核密度估计，显示完整分布形状。
# Basic violin plot / 基础小提琴图
plt.figure(figsize=(10, 6))
sns.violinplot(
    data=tips, 
    x='day', 
    y='total_bill'
)
plt.title('Bill Distribution (Violin)')
plt.show()
```

```python
# Split violin plot / 分半小提琴图
plt.figure(figsize=(12, 6))
sns.violinplot(data=tips, x='day', y='total_bill',
               hue='sex',
               split=True,               # Split by gender / 按性别分半
               palette='muted')
plt.title('Split Violin Plot by Gender')
plt.show()
```

```python
# add label for violinplot
# add mean, median and std.
fig, ax = plt.subplots(figsize=(12, 7))

# 绘制小提琴图
violin = sns.violinplot(
    data=tips, 
    x='day', 
    y='total_bill',
    hue='day', 
    ax=ax,
    inner='box', 
    palette='Set2')

# 为每个小提琴添加统计标签
categories = tips['day'].unique()

for i, category in enumerate(categories):
    data = tips[tips['day'] == category]['total_bill']
    
    # 计算统计值
    mean = data.mean()
    median = data.median()
    std = data.std()
    
    # 添加文本标签
    label_text = f'Mean: {mean:.1f}\nMedian: {median:.1f}\nStd: {std:.1f}'
    
    ax.text(i, data.max() + 1, label_text,
            ha='center', va='bottom', fontsize=9,
            bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.8, edgecolor='black'))

# 添加解释性注释
ax.annotate('White dot = Median\nBlack bar = IQR (Q1-Q3)\nThin line = Whiskers', 
            xy=(0.95, 0.95), xycoords='axes fraction',
            fontsize=10, ha='right', va='top',
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

plt.title('Violin Plot with Statistics', fontsize=14)
plt.ylim(0, tips['total_bill'].max() + 8)
plt.tight_layout()
plt.show()
```

```python
# add label for violinplot
# add detail explanation.
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# ========== for left chart. ==========
ax1 = axes[0]

sns.violinplot(
    data=tips, 
    x='day', 
    y='total_bill', 
    ax=ax1, 
    hue='day',                   
    palette='muted',
    legend=False,                
    inner='box'
)
ax1.set_title('Standard Violin Plot', fontsize=12)

# explanation
explanation = """
Violin Plot Explanation:

* Width: Frequency density (wider = more data)
* White dot: Median
* Thick bar: Interquartile Range (IQR)
* Thin line: Whiskers (1.5xIQR)
* Shape: Distribution shape
  - Symmetric: Normal distribution
  - Skewed: Uneven distribution
"""

ax1.text(1.5, tips['total_bill'].max() * 0.6, explanation,
         fontsize=9, ha='center', va='center',
         bbox=dict(boxstyle='round,pad=1', facecolor='lightyellow', alpha=0.9),
         family='monospace') 

# ========== right chart. ==========
ax2 = axes[1]

# only select one day' data for explain.
day_data = tips[tips['day'] == 'Sun']['total_bill']
parts = ax2.violinplot([day_data], positions=[0], showmeans=True, showmedians=True)

# define color
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_alpha(0.7)

# add label
stats = {
    'Min': day_data.min(),
    'Q1 (25%)': day_data.quantile(0.25),
    'Median (50%)': day_data.median(),
    'Mean': day_data.mean(),
    'Q3 (75%)': day_data.quantile(0.75),
    'Max': day_data.max()
}

y_positions = list(stats.values())
labels = list(stats.keys())

for y, label in zip(y_positions, labels):
    ax2.annotate(f'{label}: {y:.2f}', 
                 xy=(0, y), xytext=(0.3, y),
                 arrowprops=dict(arrowstyle='->', color='red', lw=1.5),
                 fontsize=10, color='darkred', fontweight='bold',
                 va='center')

ax2.set_xlim(-0.5, 0.8)
ax2.set_ylim(day_data.min() - 2, day_data.max() + 2)
ax2.set_xticks([0])
ax2.set_xticklabels(['Sunday'])
ax2.set_title('Detailed Annotation', fontsize=12)  # 英文
ax2.set_ylabel('Total Bill ($)')

plt.suptitle('Violin Plot Guide', fontsize=14)  # 英文
plt.tight_layout()
plt.show()
```

```python
# add label for violin 
fig, ax = plt.subplots(figsize=(14, 8))

# 分半小提琴图
sns.violinplot(
    data=tips, 
    x='day', 
    y='total_bill', 
    hue='sex',
    split=True, 
    palette={'Male': 'skyblue', 'Female': 'lightcoral'},
    inner='quartile', ax=ax
)

# 添加图例说明
legend_text = """
Split Violin Interpretation:

Left side (Blue): Male
Right side(Red): Female

Inner lines:
  - Solid line: Median
  - Dashed lines: Quartiles
"""

# 放置说明框
ax.text(0.02, 0.98, legend_text, transform=ax.transAxes,
        fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.9))

# 为每个分组添加统计对比
categories = tips['day'].unique()

for i, day in enumerate(categories):
    male_data = tips[(tips['day'] == day) & (tips['sex'] == 'Male')]['total_bill']
    female_data = tips[(tips['day'] == day) & (tips['sex'] == 'Female')]['total_bill']
    
    male_mean = male_data.mean()
    female_mean = female_data.mean()
    
    # 添加均值标签
    ax.text(i - 0.15, male_mean, f'male mean:\n{male_mean:.1f}', 
            ha='center', va='center', fontsize=9, color='blue', fontweight='bold')
    ax.text(i + 0.15, female_mean, f'female mean\n{female_mean:.1f}', 
            ha='center', va='center', fontsize=9, color='red', fontweight='bold')

ax.set_title('Split Violin: Gender Comparison', fontsize=14)
plt.tight_layout()
plt.show()
```
